```python
import NNLib.linear_models as linear_models
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("../Datasets/Housing/Housing.csv")
price = df['price']
area = df['area']
priceArr = np.array(price)
areaArr = np.array(area)
```

```python
linear_regressor_sgd = linear_models.LinearRegression(1e-10,10000,"sgd")
linear_regressor_sgd.fit(areaArr,priceArr)
print(linear_regressor_sgd.get_params())
print(linear_regressor_sgd.predict(740))
linear_regressor_sgd.plot()
```

```python
linear_regressor_gd = linear_models.LinearRegression(1e-10,10000,"gd")
linear_regressor_gd.fit(areaArr,priceArr)
print(linear_regressor_gd.get_params())
print(linear_regressor_gd.predict(740))
linear_regressor_gd.plot()
```
